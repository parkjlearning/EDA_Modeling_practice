### Feature engineering for df_user

```python
# Convert the string-typed 'creation_time' to datetime
df_user['creation_time'] = pd.to_datetime(df_user['creation_time'])

# Scale datetime variables
scaler = MinMaxScaler()
df_user['creation_time'] = scaler.fit_transform(df_user[['creation_time']])
df_user['last_session_creation_time'] = scaler.fit_transform(df_user[['last_session_creation_time']])

# Convert the 'org_id' to string, so that it can be further converted to a dummy variable
df_user['org_id'] = df_user['org_id'].apply(str)
```

#### 1.2 Load and preprocess user engagement data (takehome_user_engagement.csv)

```python
# Load user_engagement data
df_engage = pd.read_csv("takehome_user_engagement.csv", encoding='latin-1')
df_engage.head()
```

```python
df_engage.info()
```

```python
df_engage.describe()
```

Note that the variable 'visited' has one single value (1). 

```python
# Convert the string-typed time_stamp to datetime
df_engage['time_stamp'] = pd.to_datetime(df_engage['time_stamp'])
```

```python
# This function determines whether a user was adopted or not as defined 
def is_adopted(df, days=7): 
    
    timelimit = timedelta(days=days)
    
    user_l = []
    adopted_l = []
    
    for user in set(df['user_id']): 
        df_sort = df[df['user_id']==user].sort_values(by=['time_stamp'])
        timestamps = list(df_sort['time_stamp'])
        count, adopted = 0, 0
        
        while (len(timestamps)>2) and ((count+2)<len(timestamps)) and adopted == 0:
            if timestamps[count+2]-timestamps[count] < timelimit:
                adopted = 1
            else:
                count += 1
        
        user_l.append(user)
        adopted_l.append(adopted)
        
    df_adopt = pd.DataFrame({'user_id':user_l, 'adopted':adopted_l})     
            
    return df_adopt
```

```python
# Identify user adoption and create a new dataframe
df_adopt = is_adopted(df_engage)
```

```python
# Merge df_adopt with df_user
df_merge = pd.merge(df_user, df_adopt, left_on='object_id', right_on='user_id')
```

### 2. EDA

```python
plt.style.use('ggplot')
sns.set_context("notebook")
```

```python
# Bar plot for discrete variables
def count_bar_for_discretes(df, x_col, y_col, std_range=0):
    if std_range != 0:
        mean = np.mean(df[x_col]) 
        stdev = np.std(df[x_col])
        df = df[df[x_col] < mean + (std_range * stdev)]

    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,5))
    sns.countplot(x=x_col, hue=y_col, data=df, ax=ax[0])
    ax[0].set_title('countplot by ' + x_col, fontsize=16)
    ax[0].set_xticklabels(ax[0].get_xticklabels(),rotation=45)
    sns.barplot(x=x_col, y=y_col, data=df, ax=ax[1])
    ax[1].set_title(y_col + ' rate by ' + x_col, fontsize=16)
    ax[1].set_xticklabels(ax[1].get_xticklabels(),rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Histogram density plot for continuous variables
def histogram_density_groupby_mean(df, x_col, y_col, bin_0=30, bin_1=20, std_range=0):
    if std_range != 0:
        mean = np.mean(df[x_col]) 
        stdev = np.std(df[x_col])
        df = df[df[x_col] < mean + (std_range * stdev)]
    
    grouped = df[[x_col, y_col]].groupby(x_col).mean().reset_index()
    hist_kws={'histtype': 'bar', 'edgecolor':'black', 'alpha': 0.2}

    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))

    sns.histplot(data=df[df[y_col] == 0][x_col], ax=ax[0], stat='density', kde=True, 
                 bins=np.histogram_bin_edges(df[df[y_col] == 0][x_col], bins=bin_0), color='blue', label='y_0')
    sns.histplot(data=df[df[y_col] == 1][x_col], ax=ax[0], stat='density', kde=True, 
                 bins=np.histogram_bin_edges(df[df[y_col] == 1][x_col], bins=bin_1), color='orange', label='y_1')
    ax[0].set_title('Density plot by ' + x_col, fontsize=16)
    handles, labels = ax[0].get_legend_handles_labels()
    ax[0].legend(handles, labels)
    
    ax[1].plot(grouped[x_col], grouped[y_col], '.-')
    ax[1].set_title('mean retention rate vs. ' + x_col, fontsize=16)
    ax[1].set_xlabel(x_col)
    ax[1].set_ylabel(y_col)
    ax[1].grid(True)
    plt.show()
```

```python
# Scatter plot for discrete variables with many elements
def scatter_groupby_mean(df, x_col, y_col): 
    grouped = df[[x_col, y_col]].groupby(x_col).mean().reset_index()
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,5))
    sns.scatterplot(x=x_col, y=y_col, hue=x_col, data=grouped)
    ax.set_title('scatterplot by ' + x_col, fontsize=16)
    ax.set_xlabel(x_col, fontsize=14)
    ax.set_ylabel(y_col, fontsize=14)
    ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0)
    ax.grid(True)
    plt.show()
```

```python
def histogram_density_binby_groupby_mean(df, x_col, y_col, binby=10, bin_0=10, bin_1=10, std_range=0):
    if std_range != 0:
        mean = np.mean(df[x_col]) 
        stdev = np.std(df[x_col])
        df = df[df[x_col] < mean + (std_range * stdev)]
    
    
    df['bin'] = pd.cut(df[x_col], binby)
    grouped = df[['bin', y_col]].groupby('bin').mean().reset_index()
    
    hist_kws={'histtype': 'bar', 'edgecolor':'black', 'alpha': 0.2}

    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))

    sns.histplot(data=df[df[y_col] == 0][x_col], ax=ax[0], stat='density', kde=True, 
                 bins=np.histogram_bin_edges(df[df[y_col] == 0][x_col], bins=bin_0), color='blue', label='y_0')
    sns.histplot(data=df[df[y_col] == 1][x_col], ax=ax[0], stat='density', kde=True, 
                 bins=np.histogram_bin_edges(df[df[y_col] == 1][x_col], bins=bin_1), color='orange', label='y_1')
    ax[0].set_title('Density plot by ' + x_col, fontsize=16)
    handles, labels = ax[0].get_legend_handles_labels()
    ax[0].legend(handles, labels)
    
    ax[1].plot(list(map(lambda x: str(x), list(grouped['bin']))), grouped[y_col], '.-')
    ax[1].set_title(y_col + ' by ' + x_col, fontsize=16)
    ax[1].set_xlabel(x_col)
    ax[1].set_xticks(ax[1].get_xticks())
    ax[1].set_xticklabels(list(map(lambda x: str(x), list(grouped['bin']))), rotation=45)
    ax[1].set_ylabel(y_col)
    ax[1].grid(True)
    plt.show()
```

### 3. Predictive modeling

#### 3.1 Hyperparameter optimization with cross validation

```python
# https://machinelearningmastery.com/extreme-gradient-boosting-ensemble-in-python/
# get a list of models to evaluate
def get_XGBoost_models():
    models = dict()
    trees = [10, 50, 100, 500, 1000]
    for n in trees: 
        models[str(n)] = XGBClassifier(n_estimators=n)
    return models
```

```python
def get_RF_models():
    models = dict()
    trees = [10, 50, 100, 500, 1000]
    for n in trees: 
        models[str(n)] = RandomForestClassifier(n_estimators=n)
    return models
```

```python
# evaluate a give model using cross-validation
# Accuracy = (TP+TN)/(TP+FP+FN+TN)
def evaluate_model(model):
	cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=45)
	scores = cross_val_score(model, X_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1)
	return scores
```

#### 3.2 Split dataset into train and test sets 

```python
df = df_merge.drop(axis=1, columns=['object_id', 'name', 'email', 'user_id', 'invited_by_user_id'])
```

```python
df = pd.get_dummies(df)
```

```python
y = df['adopted']
X = df.drop(axis=1, columns=['adopted']) 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

#### 3.3 Optimize, train and evaluate Random Forest classifier 

#### 3.3.1 Hyperparameter optimization for Random Forest classifier with grid search

```python
# get XGBoost models to evaluate
rf_models = get_RF_models()

# evalutate the models and store results
rf_results, rf_names = list(), list()
for name, model in rf_models.items(): 
    rf_scores = evaluate_model(model)
    rf_results.append(rf_scores)
    rf_names.append(name)
    print(name + ', mean: {0:.3f}, std: {1:.3f}'.format(np.mean(rf_scores), np.std(rf_scores)))
    
# plot model performance for comparison
fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.boxplot(rf_results, labels=rf_names, showmeans=True)
ax.set_title("Model Evaluation Random Forest Classifier", fontsize=16)
ax.set_xlabel("Number of Trees", fontsize=14)
ax.set_ylabel("Accuracy", fontsize=14)
plt.show()
```

#### 3.3.2 Fit, predict, evaluate Random Forest classifier

```python
# fit model no training data
rf_model = RandomForestClassifier(n_estimators=1000)
rf_model.fit(X_train, y_train)
print(rf_model)
# make predictions for test data
y_pred_rf = rf_model.predict(X_test)
rf_predictions = [round(value) for value in y_pred_rf]
```

```python
# classification report
print(classification_report(y_test, y_pred_rf))
# confusion matrix
print(confusion_matrix(y_test, y_pred_rf))
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_rf).ravel()
```

```python
importance = rf_model.feature_importances_

df_rf_importance = pd.DataFrame({'variable': X_train.columns, 'scaled_importance':importance})
df_rf_importance_top20 = df_rf_importance.sort_values(by=['scaled_importance'], ascending=False).iloc[:20]

fig, ax = plt.subplots(figsize=(10,8))
sns.barplot(x='scaled_importance', y='variable', data=df_rf_importance_top20, ax=ax)
ax.set_title('Feature importance Random Forest Model')
plt.show()
```

#### 3.4 Optimize, train and evaluate XGBoost classifier 

#### 3.4.1 Hyperparameter optimization for XGBoost classifier with grid search

```python
# get XGBoost models to evaluate
xgb_models = get_XGBoost_models()

# evalutate the models and store results
xgb_results, xgb_names = list(), list()
for name, model in xgb_models.items(): 
    xgb_scores = evaluate_model(model)
    xgb_results.append(xgb_scores)
    xgb_names.append(name)
    print(name + ', mean: {0:.3f}, std: {1:.3f}'.format(np.mean(xgb_scores), np.std(xgb_scores)))
    
# plot model performance for comparison
fig, ax = plt.subplots(1,1,figsize=(10,6))
ax.boxplot(xgb_results, labels=xgb_names, showmeans=True)
ax.set_title("Model Evaluation XGBoost Classifier", fontsize=16)
ax.set_xlabel("Number of Trees", fontsize=14)
ax.set_ylabel("Accuracy", fontsize=14)
plt.show()
```

#### 3.4.2 Fit, predict, evaluate XGBoost classifier

```python
# fit model no training data
xgb_model = XGBClassifier(n_estimators=10, use_label_encoder=False)
xgb_model.fit(X_train, y_train)
print(xgb_model)
# make predictions for test data
y_pred_xgb = xgb_model.predict(X_test)
xgb_predictions = [round(value) for value in y_pred_xgb]
```

```python
# classification report
print(classification_report(y_test, y_pred_xgb))
# confusion matrix
print(confusion_matrix(y_test, y_pred_xgb))
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_xgb).ravel()
```

```python
importance = xgb_model.feature_importances_

df_xgb_importance = pd.DataFrame({'variable': X_train.columns, 'scaled_importance':importance})
df_xgb_importance_top20 = df_xgb_importance.sort_values(by=['scaled_importance'], ascending=False).iloc[:20]

fig, ax = plt.subplots(figsize=(10,8))
sns.barplot(x='scaled_importance', y='variable', data=df_xgb_importance_top20, ax=ax)
ax.set_title('Feature importance Random Forest Model')
plt.show()
```

#### 3.6 Summary

We analyzed user profile and user engagement data, based upon which we have built predictive models that can predict user adoption. Our EDA has made interesting observations as below: 
1. People from different organizations differ in their adoption rates. 
2. There was a recent improvement in adoption rate observed, i.e., there was a significant increase in adoption rate from users who created their final sessions most recently. 

Based on these observations we could build predictive model variants - random forest, XGBoost classifiers. Both models performed with high accuracy, but the best performance was achieved by the XGBoost classifier with 10 trees. The scaled feature importance data showed that models highly relied on the 'last_session_creation_time' variable, consistent with our observation made during EDA.
